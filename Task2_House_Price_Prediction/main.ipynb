# House Price Prediction using Linear Regression

This project predicts house prices using Linear Regression based on important house features.

```python
# Importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Loading the Dataset

```python
# Loading the housing dataset

housing_df = pd.read_csv("Housing.csv")

# Displaying first 5 rows
housing_df.head()
```

## Understanding the Dataset

```python
# Checking dataset shape
housing_df.shape
```

```python
# Checking column names and data types
housing_df.info()
```

```python
# Checking missing values
housing_df.isnull().sum()
```

## Data Cleaning

```python
# Checking duplicate values

housing_df.duplicated().sum()
```

```python
# Removing duplicate values if present

housing_df = housing_df.drop_duplicates()
```

```python
# Checking dataset after cleaning

housing_df.shape
```

## Feature Selection

```python
# Selecting important features

features = housing_df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]

# Target variable
target = housing_df['price']
```

```python
# Displaying selected features

features.head()
```

```python
# Displaying target values

target.head()
```

## Splitting the Dataset

```python
# Splitting dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)
```

## Building Linear Regression Model

```python
# Creating Linear Regression model

model = LinearRegression()
```

```python
# Training the model

model.fit(X_train, y_train)
```

## Making Predictions

```python
# Predicting house prices

y_pred = model.predict(X_test)
```

```python
# Displaying predicted values

print(y_pred[:5])
```

## Model Evaluation

```python
# Calculating Mean Squared Error

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```

```python
# Calculating R2 Score

r2 = r2_score(y_test, y_pred)

print("R2 Score:", r2)
```

## Actual vs Predicted Prices

## Data Visualization

```python
# Visualizing actual vs predicted prices

plt.figure(figsize=(8,5))

plt.scatter(y_test, y_pred)

plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")

plt.title("Actual vs Predicted House Prices")

plt.show()
```
